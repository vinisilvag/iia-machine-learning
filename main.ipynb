```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb

def regression_error_metrics(y_test, y_pred):
    mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
    mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
    rmse = mean_squared_error(y_true=y_test, y_pred=y_pred, squared=False)
    r2 = r2_score(y_true=y_test, y_pred=y_pred)
      
    print("  MAE:", mae) 
    print("  MSE:", mse) 
    print("  RMSE:", rmse)
    print("  R2 Score:", r2)

    return mae, mse, rmse, r2

regression_comparison_data = []
```

```python
X = df[df.columns[~df.columns.isin(["Goal"])]]
y = df["Goal"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=444)
```

## Regressão Linear

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)

y_pred = linear_regression.predict(X_test)

print("Linear Regression:")
mae, mse, rmse, r2 = regression_error_metrics(y_test, y_pred)

regression_comparison_data.append(("linear-regression", pd.NA, pd.NA, pd.NA, pd.NA, mae, mse, rmse, r2))
```

## Ridge

```python
alphas = [0.2, 1, 5]

for alpha in alphas:
    print(f"Ridge - alpha = {alpha}:")
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train, y_train)
    
    y_pred = ridge.predict(X_test)

    mae, mse, rmse, r2 = regression_error_metrics(y_test, y_pred)
    print()

    regression_comparison_data.append(("ridge", alpha, pd.NA, pd.NA, pd.NA, mae, mse, rmse, r2))
```

## XGBoost

```python
params = [[50, 4, 0.2],
          [50, 10, 1],
          [50, 10, 0.2],
          [50, 10, 0.04],
          [200, 4, 1],
          [200, 4, 0.2],
          [200, 4, 0.04],
          [200, 10, 0.2]]

for p in params:
    n_estimators, max_depth, learning_rate = p
    print(f"XGBoost Regression - n_estimators = {n_estimators}, max_depth = {max_depth}, learning_rate = {learning_rate}:")
    
    # por padrão minimiza o MSE
    xgb_regression = xgb.XGBRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=444
    )
    
    xgb_regression.fit(X_train, y_train)
    y_pred = xgb_regression.predict(X_test)

    mae, mse, rmse, r2 = regression_error_metrics(y_test, y_pred)
    print()

    regression_comparison_data.append(("xgboost", pd.NA, n_estimators, max_depth, learning_rate, mae, mse, rmse, r2))
```

```python
regression_comparison_df = pd.DataFrame(np.array(regression_comparison_data), columns=["strategy", "alpha", "n_estimators", "max_depth",
                                                              "learning_rate", "MAE", "MSE", "RMSE", "R2"])
regression_comparison_df
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
import xgboost as xgb

def classifier_error_metrics(y_test, y_pred, average):
    accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
    recall = recall_score(y_true=y_test, y_pred=y_pred, average=average, zero_division=0)
    precision = precision_score(y_true=y_test, y_pred=y_pred, average=average, zero_division=0)
    f1 = f1_score(y_true=y_test, y_pred=y_pred, average=average, zero_division=0)
      
    print("  Accuracy:", accuracy) 
    print("  Recall:", recall) 
    print("  Precision:", precision)
    print("  F1 Score:", f1)

    return accuracy, recall, precision, f1

classifier_comparison_data = []
```

```python
# np.inf no lugar de inf por causa de erros de execução
classes = {0.0: (-np.inf, -0.0025),
           1.0: (-0.0025, -0.0020),
           2.0: (-0.0020, -0.0015),
           3.0: (-0.0015, -0.0010),
           4.0: (-0.0010, -0.0005),
           5.0: (-0.0005, np.inf)}

bins = []
for values in list(classes.values()):
    bins.append(values[0])
    bins.append(values[1])
bins = list(dict.fromkeys(bins))
labels = tuple(classes.keys())

df["DiscreteGoal"] = pd.cut(df["Goal"], bins=bins, labels=labels)
```

```python
def display_frames_as_gif(frames):
    plt.ioff()
    aspect = 72
    fig, ax = plt.subplots(figsize=(frames[0].shape[1] / aspect, frames[0].shape[0] / aspect), dpi=aspect)
    patch = ax.imshow(frames[0])
    _ = plt.axis('off')

    def animate(i):
        patch.set_data(frames[i])

    anim = animation.FuncAnimation(fig, animate, frames=len(frames), interval=50, )
    return anim
```

```python
env = gym.make('CartPole-v1', render_mode = 'rgb_array')

# Q-learning parameters
alpha = 0.1  
gamma = 0.99  

epsilon_zero = 0.9  # valor inicial de epsilon
epsilon_min = 0.0  # valor final de epsilon
episodes = 500  # Numero maximo de episodios do treinamento. MAXIMO PERMITIDO 250_000
min_epsilon_episode = 380 # Episodio que o epsilon mínimo é atingido

max_steps = 500  # Número máximo de passos no treinamento. Fixo! 
epsilon = epsilon_zero
epsilon_decay = (epsilon_min - epsilon) / min_epsilon_episode  # Decaimento Linear do Epsilon, PERMITIDO ALTERAÇÕES
```

```python
#TODO
# Faça um plot da variação de epsilon durante o decorrer de 'episodes'

x = np.arange(1, episodes + 1)
```

```python
# Define the bins for each state variable
n_bins = 10 # Escolha um número de estados para cada variável
bins = [
    np.linspace(-4.8, 4.8, n_bins), #Cart Position
    np.linspace(-4, 4, n_bins), #Cart Velocity
    np.linspace(-0.418, 0.418, n_bins), #Pole Angle
    np.linspace(-4, 4, n_bins) #Pole Angular Velocity
]

def discretize_state(state, bins):
    """ 
        As variaveis do CartPole originalmente são contínuas. Discretize.
        Implemente a função discretize_state(state, bins) -> state_discretized
            state: list | tuple = Estado do vagão
            bins: Lista com as discretizações

            return state_discretized: tuple = Estados discretizados
    """
    
    return tuple(np.digitize(state[i], bins[i]) for i in range(4))

# Inicialização da Q-table
state_space_size = tuple([len(bins[i]) + 1 for i in range(len(bins))])
action_space_size = env.action_space.n
Q = np.zeros(state_space_size + (action_space_size,))
```

```python
def choose_action(state, Q, epsilon):
    """
        Implemente a função choose_action(state, Q, epsilon)
        Escolha de forma probabilística conforme epsilon entre uma ação aleatória ou uma ação ótima
    """
    
    if np.random.rand() < epsilon:
        return np.random.randint(0, 2)
    else:
        return np.argmax(Q[state])

def update_q_table(Q, state, action, reward, next_state, alpha, gamma):
    """
        Faça a atualização da Q-table conforme formulação do RL
    """
    # Q(state, action) = Q(s, action) + alpha * (r + gamma * Q(next_state, action) - Q(state, action))

    # repetindo a mesma ação realizada anteriormente
    # Q[state, action] = Q[state, action] + alpha * (reward + gamma * Q[next_state, action] - Q[state, action])

    # escolhendo a melhor ação possível no novo estado
    Q[state, action] = Q[state, action] + alpha * (reward + (gamma * max(Q[next_state])) - Q[state, action])
```

```python
#Treinamento da Q-table
for episode in range(episodes):
    state, _ = env.reset()
    state = discretize_state(state, bins)
    total_reward = 0

    for step in range(max_steps):
        action = choose_action(state, Q, epsilon)
        next_state, reward, done, truncated, _ = env.step(action)
        next_state = discretize_state(next_state, bins)

        update_q_table(Q, state, action, reward, next_state, alpha, gamma)

        state = next_state
        total_reward += reward

        if done:
            break

    # Decay epsilon
    if epsilon > epsilon_min:
        epsilon = epsilon_decay * episode + epsilon_zero

    print(f"Episode {episode + 1}: Total Reward = {total_reward}")

print("Training finished.\n")
```

```python
frames = []

state, _ = env.reset()
state = discretize_state(state, bins)
total_reward = 0

for step in range(max_steps):
    frames.append(env.render())

    action = np.argmax(Q[state])
    next_state, reward, done, truncated, _  = env.step(action)
    next_state = discretize_state(next_state, bins)

    state = next_state
    total_reward += reward
    
    if done:
        break

print(f"Test Episode {episode + 1}: Total Reward = {total_reward}")
```
