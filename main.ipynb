## I. Porque os atributos Alpha, Se e [SeTime1 : SeTime14] não contribuem em conjunto para um modelo?

```python
filtered = df[["Alpha", "Se", *[f"SeTime{i}" for i in range(1, 15)]]]
corr = filtered.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True)
plt.show()
```

## **Resposta:** Porque eles possuem uma alta correlação entre si. Idealmente, queremos escolher apenas um desses atributos, reduzindo o número de dimensões dos dados, o que alivia a maldição da dimensionalidade. Além disso, a escolha de atributos altamente correlacionados prejudica o desempenho dos classificadores/regressores, que podem apresentar resultados inferiores aos obtidos caso a *feature selection* seja feita apropriadamente, e dificulta a interpretação e análise da contribuição de cada variável para o preditor implementado.

# VIII. Elabore os seguintes modelos de regressão:

| Modelo | Espaço Hiperparâmetro | 
| :- | :-: |
| Regressão Linear | - |
| Ridge | alpha: {0.2, 1, 5} |
| XGboost | **instrucoes.txt** |

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb

def regression_error_metrics(y_test, y_pred):
    mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
    mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
    rmse = mean_squared_error(y_true=y_test, y_pred=y_pred, squared=False)
    r2 = r2_score(y_true=y_test, y_pred=y_pred)
      
    print("  MAE:", mae) 
    print("  MSE:", mse) 
    print("  RMSE:", rmse)
    print("  R2 Score:", r2)

    return mae, mse, rmse, r2

regression_comparison_data = []
```

```python
X = np.array(df[df.columns[~df.columns.isin(["Goal"])]]) 
y = np.array(df["Goal"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=444)
```

## Regressão Linear

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)

y_pred = linear_regression.predict(X_test)

print("Linear Regression:")
mae, mse, rmse, r2 = regression_error_metrics(y_test, y_pred)

regression_comparison_data.append(("linear-regression", pd.NA, pd.NA, pd.NA, pd.NA, mae, mse, rmse, r2))
```

## Ridge

```python
alphas = [0.2, 1, 5]

for alpha in alphas:
    print(f"Ridge - alpha = {alpha}:")
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train, y_train)
    
    y_pred = ridge.predict(X_test)

    mae, mse, rmse, r2 = regression_error_metrics(y_test, y_pred)
    print()

    regression_comparison_data.append(("ridge", alpha, pd.NA, pd.NA, pd.NA, mae, mse, rmse, r2))
```

## XGBoost

```python
params = [[50, 4, 0.2],
          [50, 10, 1],
          [50, 10, 0.2],
          [50, 10, 0.04],
          [200, 4, 1],
          [200, 4, 0.2],
          [200, 4, 0.04],
          [200, 10, 0.2]]

for p in params:
    n_estimators, max_depth, learning_rate = p
    print(f"XGBoost Regression - n_estimators = {n_estimators}, max_depth = {max_depth}, learning_rate = {learning_rate}:")
    
    # por padrão minimiza o MSE
    xgb_regression = xgb.XGBRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=444
    )
    
    xgb_regression.fit(X_train, y_train)
    y_pred = xgb_regression.predict(X_test)

    mae, mse, rmse, r2 = regression_error_metrics(y_test, y_pred)
    print()

    regression_comparison_data.append(("xgboost", pd.NA, n_estimators, max_depth, learning_rate, mae, mse, rmse, r2))
```

```python
regression_comparison_df = pd.DataFrame(np.array(regression_comparison_data), columns=["strategy", "alpha", "n_estimators", "max_depth",
                                                              "learning_rate", "MAE", "MSE", "RMSE", "R2"])
regression_comparison_df
```

# I. Elabore os seguintes modelos de classificação:

| Modelo | Espaço Hiperparâmetro | 
| :- | :-: |
| Regressão Logística | - |
| Naive Bayes (Gaussian) | - |
| Árvores de Decisão | max_depth: {X, Y} |
| K-NN | n_neighbors: {W, Z} |
| XGboost | **instrucoes.txt** |

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
import xgboost as xgb

def classifier_error_metrics(y_test, y_pred, average):
    accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
    recall = recall_score(y_true=y_test, y_pred=y_pred, average=average, zero_division=0)
    precision = precision_score(y_true=y_test, y_pred=y_pred, average=average, zero_division=0)
    f1 = f1_score(y_true=y_test, y_pred=y_pred, average=average, zero_division=0)
      
    print("  Accuracy:", accuracy) 
    print("  Recall:", recall) 
    print("  Precision:", precision)
    print("  F1 Score:", f1)

    return accuracy, recall, precision, f1

classifier_comparison_data = []
```

```python
# np.inf no lugar de inf por causa de erros de execução
classes = {0.0: (-np.inf, -0.0025),
           1.0: (-0.0025, -0.0020),
           2.0: (-0.0020, -0.0015),
           3.0: (-0.0015, -0.0010),
           4.0: (-0.0010, -0.0005),
           5.0: (-0.0005, np.inf)}

bins = []
for values in list(classes.values()):
    bins.append(values[0])
    bins.append(values[1])
bins = list(dict.fromkeys(bins))
labels = tuple(classes.keys())

df["DiscreteGoal"] = pd.cut(df["Goal"], bins=bins, labels=labels)
```

```python
plt.hist(df["DiscreteGoal"])
```

```python
X = np.array(df[df.columns[~df.columns.isin(["DiscreteGoal", "Goal"])]])
y = np.array(df["DiscreteGoal"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=444)
```

## Regressão Logística

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)

y_pred = logistic_regression.predict(X_test)

print("Logistic Regression:")
accuracy, recall, precision, f1 = classifier_error_metrics(y_test, y_pred, "weighted")

classifier_comparison_data.append(("logistic-regression", pd.NA, pd.NA, pd.NA, pd.NA, pd.NA, accuracy, recall, precision, f1))
```

## Naive Bayes (Gaussian)

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)

y_pred = naive_bayes.predict(X_test)

print("Naive Bayes:")
accuracy, recall, precision, f1 = classifier_error_metrics(y_test, y_pred, "weighted")

classifier_comparison_data.append(("naive-bayes", pd.NA, pd.NA, pd.NA, pd.NA, pd.NA, accuracy, recall, precision, f1))
```

## Árvores de Decisão

```python
depths = np.arange(2, 52, 2)

for max_depth in depths:
    print(f"Decision Tree - max_depth = {max_depth}:")
    
    decision_tree = DecisionTreeClassifier(max_depth=depth)
    decision_tree.fit(X_train, y_train)
    
    y_pred = decision_tree.predict(X_test)
    
    accuracy, recall, precision, f1 = classifier_error_metrics(y_test, y_pred, "weighted")
    print()

    classifier_comparison_data.append(("decision-tree", max_depth, pd.NA, pd.NA, pd.NA, pd.NA, accuracy, recall, precision, f1))
```

## K-NN

```python
neighbors = np.arange(2, 52, 2)

for n_neighbors in neighbors:
    print(f"K-NN - n_neighbors = {n_neighbors}:")
    
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_test)
    
    accuracy, recall, precision, f1 = classifier_error_metrics(y_test, y_pred, "weighted")
    print()

    classifier_comparison_data.append(("knn", pd.NA, n_neighbors, pd.NA, pd.NA, pd.NA, accuracy, recall, precision, f1))
```

## XGBoost

```python
params = [[50, 4, 0.2],
          [50, 10, 1],
          [50, 10, 0.2],
          [50, 10, 0.04],
          [200, 4, 1],
          [200, 4, 0.2],
          [200, 4, 0.04],
          [200, 10, 0.2]]

for p in params:
    n_estimators, max_depth, learning_rate = p
    print(f"XGBoost Classifier - n_estimators = {n_estimators}, max_depth = {max_depth}, learning_rate = {learning_rate}:")
    
    # por padrão minimiza o MSE
    xgb_classifier = xgb.XGBClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=444
    )
    
    xgb_classifier.fit(X_train, y_train)
    y_pred = xgb_classifier.predict(X_test)

    accuracy, recall, precision, f1 = classifier_error_metrics(y_test, y_pred, "weighted")
    print()

    classifier_comparison_data.append(("xgboost", pd.NA, pd.NA, n_estimators, max_depth, learning_rate, accuracy, recall, precision, f1))
```

```python
classifier_comparison_df = pd.DataFrame(np.array(classifier_comparison_data), columns=["strategy", "dt_max_depth", "n_neighbors",
                                                                                       "n_estimators", "max_depth", "learning_rate", "accuracy",
                                                                                       "recall", "precision", "F1"])
classifier_comparison_df
```
